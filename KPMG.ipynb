```python
def clean_df(demo_df):
    demo_df = demo_df.replace(to_replace='New South Wales', value='NSW')
    demo_df = demo_df.replace(to_replace='Victoria', value='VIC')
    demo_df = demo_df.replace(to_replace='F', value='Female')
    demo_df = demo_df.replace(to_replace='M', value='Male')
    demo_df = demo_df.replace(to_replace='Femal', value='Female')
    demo_df = demo_df.replace(to_replace=pd.to_datetime('1843-12-21'), value=pd.to_datetime('1943-12-21'))
    demo_df = demo_df.replace(to_replace='No', value=0)
    demo_df = demo_df.replace(to_replace='Yes', value=1)
    demo_df = demo_df.replace(to_replace='Mass Customer', value=1000)
    demo_df = demo_df.replace(to_replace='Affluent Customer', value=40000)
    demo_df = demo_df.replace(to_replace='High Net Worth', value=20000)
    demo_df['age'] = pd.to_datetime(date.today()) - demo_df['DOB']
#     demo_df = demo_df.dropna()
    demo_df['age'] = pd.to_numeric(demo_df['age'].dt.days, downcast='integer')
    demo_df['age'] = demo_df['age'].fillna((demo_df.age.mean()))
    print(len(demo_df))
    return demo_df




```
